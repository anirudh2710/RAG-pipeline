```python
sample_text={ "../data/text_files/python_intro.txt": """ Subject: Project AETHER - Phase 4 Implementation Guidelines

Project AETHER was initiated on October 14, 2023, under the supervision of Dr. Aris Thorne. The primary goal is to reduce latent heat signatures in server farm cooling units by 22%.

Key Protocols:

The coolant mixture must maintain a ratio of 4:1 (distilled water to Glycol-7).

Maintenance is scheduled bi-weekly on Tuesdays at 03:00 UTC.

The emergency override code for Sector 7 is 9921-Alpha.

Failure to adhere to the pressure threshold of 45 PSI will trigger an automatic shutdown of the primary turbine.""",
"../data/text_files/python_intro1.txt": """Memo: Corporate Policy on Remote Work (Effective Jan 2026)

While the company values flexibility, the "Core Collaboration Hours" (CCH) have been updated. Employees must be online between 10:00 AM and 2:00 PM EST. However, this does not apply to the DevOps team, who follow a "Follow-the-Sun" model.

Paradoxically, while productivity increased by 12% during the pilot program, employee satisfaction scores dipped slightly due to "digital fatigue." To combat this, the "No-Meeting Fridays" initiative was introduced, though it is often bypassed during the final week of a fiscal quarter."""
}
```

```python
for filepath, content in sample_text.items():
    with open(filepath, 'w', encoding="utf-8") as f:
        f.write(content)

print('Sample text file created!')
```

```python
### textloader
from langchain_community.document_loaders import TextLoader
loader = TextLoader("../data/text_files/python_intro.txt", encoding="utf-8")
doc1 = loader.load()
print(doc1)
```

```python
### pdfloader
from langchain_community.document_loaders import PyPDFLoader, PyMuPDFLoader, DirectoryLoader

loader = PyPDFLoader(
    file_path="../data/pdf_files/Satwave_Brochure.pdf",
    headers=None,
    extract_images=False
)
pdf1 = loader.load()
print(pdf1)

loader2 = DirectoryLoader(
    "../data/pdf_files",
    glob="**/*.pdf",
    loader_cls=PyMuPDFLoader,
    show_progress=False
)
pdf2 = loader2.load()
print(pdf2)
```
